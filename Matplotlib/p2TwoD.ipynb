```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------- Step 1: Prepare Data ----------------
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales = np.random.randint(200, 500, size=12)

products = ["Laptop", "Tablet", "Phone", "Desktop"]
product_sales = [250, 150, 300, 100]

ages = np.random.randint(18, 60, size=100)

hours = [2, 3, 4, 5, 6, 7, 8]
marks = [50, 55, 60, 65, 70, 75, 85]

categories = ["Rent", "Food", "Transport", "Entertainment", "Savings"]
expenses = [5000, 2000, 1200, 800, 1500]

days = [1, 2, 3, 4, 5]
google = [50, 60, 70, 80, 90]
facebook = [30, 40, 50, 60, 70]
twitter = [20, 30, 40, 50, 60]

salaries = np.random.normal(50000, 8000, 200)

# ---------------- Step 2: Create Subplots ----------------
fig, axs = plt.subplots(4, 2, figsize=(14, 16))
fig.suptitle("📊 2D Plots in Matplotlib", fontsize=20)

# ---------------- Line Plot ----------------
axs[0, 0].plot(months, sales, marker='o', linestyle='--', color='blue', label="Sales")
axs[0, 0].set_title("Line Plot - Monthly Sales")
axs[0, 0].set_xlabel("Months")
axs[0, 0].set_ylabel("Sales")
axs[0, 0].legend()
axs[0, 0].grid(True)

# ---------------- Bar Chart ----------------
axs[0, 1].bar(products, product_sales, color=['red','blue','green','orange'])
axs[0, 1].set_title("Bar Plot - Product Sales")
axs[0, 1].set_xlabel("Products")
axs[0, 1].set_ylabel("Units Sold")

# ---------------- Histogram ----------------
axs[1, 0].hist(ages, bins=8, color='purple', edgecolor='black')
axs[1, 0].set_title("Histogram - Customer Ages")
axs[1, 0].set_xlabel("Age")
axs[1, 0].set_ylabel("Number of Customers")

# ---------------- Scatter Plot ----------------
axs[1, 1].scatter(hours, marks, color="red", marker="*")
axs[1, 1].set_title("Scatter Plot - Study Hours vs Marks")
axs[1, 1].set_xlabel("Hours")
axs[1, 1].set_ylabel("Marks")

# ---------------- Pie Chart ----------------
axs[2, 0].pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90)
axs[2, 0].set_title("Pie Chart - Monthly Budget")

# ---------------- Stack Plot ----------------
axs[2, 1].stackplot(days, google, facebook, twitter, 
                    labels=['Google','Facebook','Twitter'],
                    colors=['red','blue','green'])
axs[2, 1].legend(loc='upper left')
axs[2, 1].set_title("Stack Plot - Website Visitors")

# ---------------- Box Plot ----------------
axs[3, 0].boxplot(salaries, vert=True, patch_artist=True)
axs[3, 0].set_title("Box Plot - Salary Distribution")
axs[3, 0].set_ylabel("Salary")

# ---------------- Empty Plot ----------------
axs[3, 1].text(0.5, 0.5, "All 2D plots demo", fontsize=16, ha='center')
axs[3, 1].axis('off')  # Remove axes

# ---------------- Step 3: Show Figure ----------------
plt.tight_layout()
plt.show()
```
