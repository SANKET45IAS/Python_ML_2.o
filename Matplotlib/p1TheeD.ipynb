```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # For 3D plots

# ---------------- Step 1: Prepare Data ----------------
# Monthly sales data
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales = np.random.randint(200, 500, size=12)

# Product sales
products = ["Laptop", "Tablet", "Phone", "Desktop"]
product_sales = [250, 150, 300, 100]

# Customer ages
ages = np.random.randint(18, 60, size=100)

# Study data
hours = [2, 3, 4, 5, 6, 7, 8]
marks = [50, 55, 60, 65, 70, 75, 85]

# Budget data
categories = ["Rent", "Food", "Transport", "Entertainment", "Savings"]
expenses = [5000, 2000, 1200, 800, 1500]

# Website visitors
days = [1, 2, 3, 4, 5]
google = [50, 60, 70, 80, 90]
facebook = [30, 40, 50, 60, 70]
twitter = [20, 30, 40, 50, 60]

# Salary distribution
salaries = np.random.normal(50000, 8000, 200)

# 3D profit data
X = np.arange(1, 13)
Y = np.random.randint(100, 200, size=12)
Z = np.random.randint(50, 100, size=12)

# ---------------- Step 2: Create Subplots ----------------
fig, axs = plt.subplots(4, 2, figsize=(14, 16))  # 4x2 grid
fig.suptitle(" Company Data Analysis Dashboard", fontsize=20)

# ---------------- Line Plot ----------------
axs[0, 0].plot(months, sales, marker='o', linestyle='--', color='blue', label="Sales")
axs[0, 0].set_title("Monthly Sales Trend")
axs[0, 0].set_xlabel("Months")
axs[0, 0].set_ylabel("Sales (units)")
axs[0, 0].legend()
axs[0, 0].grid(True)

# ---------------- Bar Chart ----------------
axs[0, 1].bar(products, product_sales, color=['red','blue','green','orange'])
axs[0, 1].set_title("Product Sales Comparison")
axs[0, 1].set_xlabel("Products")
axs[0, 1].set_ylabel("Units Sold")

# ---------------- Histogram ----------------
axs[1, 0].hist(ages, bins=8, color='purple', edgecolor='black')
axs[1, 0].set_title("Customer Age Distribution")
axs[1, 0].set_xlabel("Age")
axs[1, 0].set_ylabel("Count")

# ---------------- Scatter Plot ----------------
axs[1, 1].scatter(hours, marks, color="red", marker="*")
axs[1, 1].set_title("Study Hours vs Marks")
axs[1, 1].set_xlabel("Hours of Study")
axs[1, 1].set_ylabel("Marks")

# ---------------- Pie Chart ----------------
axs[2, 0].pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90)
axs[2, 0].set_title("Monthly Budget Breakdown")

# ---------------- Stack Plot ----------------
axs[2, 1].stackplot(days, google, facebook, twitter, 
                    labels=['Google','Facebook','Twitter'],
                    colors=['red','blue','green'])
axs[2, 1].legend(loc='upper left')
axs[2, 1].set_title("Website Visitors by Source")

# ---------------- Box Plot ----------------
axs[3, 0].boxplot(salaries, vert=True, patch_artist=True)
axs[3, 0].set_title("Employee Salary Distribution")
axs[3, 0].set_ylabel("Salary")

# ---------------- 3D Plot ----------------
ax3d = fig.add_subplot(4, 2, 8, projection='3d')
ax3d.plot(X, Y, Z, marker='o')
ax3d.set_title("3D Profit Growth")
ax3d.set_xlabel("Month")
ax3d.set_ylabel("Revenue")
ax3d.set_zlabel("Profit")

# ---------------- Show & Save ----------------
plt.tight_layout()
plt.savefig("company_dashboard.png", dpi=300)  # Save figure
plt.show()
```
